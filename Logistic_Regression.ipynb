```python
# Logistic regression
# Our aim is to solve classig=fication problem.

# In precision, false positive is important.
# In Recall, false neagtive is important.

# Balanced dataset= accuracy
# Imbalanced dataset= precision
```

```python
from sklearn.datasets import load_iris
```

```python
ds=load_iris()
ds
```

```python
print(ds.DESCR)
```

```python
ds.keys()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ds.feature_names
```

```python
# Coverting dataset into dataframe.

df=pd.DataFrame(ds.data,columns=ds.feature_names)
```

```python
df.head()
```

```python
df["target"]=ds.target
```

```python
df.head()
```

```python
ds.target
```

```python
# Binary Classification

df["target"]!=2
```

```python
df_copy=df[df["target"]!=2]
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
# Dependent and independent features.

x=df_copy.iloc[:,:-1]
y=df_copy.iloc[:,-1]
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression

lg=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
# Fitting data

lg.fit(x_train,y_train)
```

```python
# Prediction

y_pred=lg.predict(x_test)
y_pred
```

```python
lg.predict_proba(x_test)
```

```python
# Calculating metrics

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
# hyperparameter tuning

#GridSearchCV

from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')

parameters = {'penalty':['l1','l2','elasticnet',None],
              'C' : [1,10,20,30],
              'max_iter' : [100, 200, 300]
              }

clf = GridSearchCV(lg, param_grid = parameters)
```

```python
# Splitting of train data to validation data
clf.fit(x_train, y_train)
```

```python
clf.best_params_
```

```python
classifier = LogisticRegression(C = 1, penalty = 'l2',max_iter = 100)
```

```python
classifier.fit(x_train,y_train)
```

```python
# Prediction

y_pred = classifier.predict(x_test)

print(confusion_matrix(y_pred, y_test))
print(accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
# RandomizedSearchCV

from sklearn.model_selection import RandomizedSearchCV

random_clf = RandomizedSearchCV(LogisticRegression(), param_distributions=parameters)

random_clf.fit(x_train,y_train)

random_clf.best_params_

classifier = LogisticRegression(C = 10, penalty = None, max_iter = 100)

classifier.fit(x_train, y_train)

## Prediction
y_pred = classifier.predict(x_test)
print(confusion_matrix(y_pred, y_test))
print(accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
# Import pickle for saving and loading the model

import pickle

# Filepath for saving the model
pickle_file_path = 'logistic.pkl'

# Save the trained model to a pickle file
with open(pickle_file_path, 'wb') as file:
    pickle.dump(classifier, file)

print(f'Model saved to {pickle_file_path}')

# Load the model from the pickle file
with open(pickle_file_path, 'rb') as file:
    loaded_model = pickle.load(file)

# Test the loaded model with predictions
y_pred_loaded = loaded_model.predict(x_test)
print(y_pred_loaded)
```
